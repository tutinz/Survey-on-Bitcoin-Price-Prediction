## Bitcoin price evolution

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical data for Bitcoin (BTC-USD)
BTC_Ticker = yf.Ticker("BTC-USD")
BTC_Data = BTC_Ticker.history(period="max")

# Calculate volatility (standard deviation of daily returns)
BTC_Data['Volatility'] = BTC_Data['Close'].pct_change().rolling(window=30).std()

# Plot time series data and its volatility
plt.figure(figsize=(12, 8))

# Plot time series data
plt.subplot(2, 1, 1)
plt.plot(BTC_Data.index, BTC_Data['Close'], color='blue')
plt.title('Bitcoin (BTC-USD) Price Evolution', fontsize=17)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Price (USD)', fontsize=15)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Import pandas

# Fetch historical data for Bitcoin (BTC-USD)
BTC_Ticker = yf.Ticker("BTC-USD")
BTC_Data = BTC_Ticker.history(period="max")

# Calculate volatility (standard deviation of daily returns)
BTC_Data['Volatility'] = BTC_Data['Close'].pct_change().rolling(window=30).std()

# Calculate log prices
BTC_Data['Log_Close'] = np.log(BTC_Data['Close'])

# Halving dates (approximate)
halving_dates = [
#    "2012-11-28",  # 1st halving
    "2016-07-09",  # 2nd halving
    "2020-05-11",  # 3rd halving
    "2024-04-01"   # Estimated 4th halving
]

# Plot time series data, log prices, and volatility
plt.figure(figsize=(12, 10))


# Plot log prices
plt.subplot(3, 1, 2)
plt.plot(BTC_Data.index, BTC_Data['Log_Close'], color='green')
plt.title('Bitcoin (BTC-USD) Log Price Evolution', fontsize=17)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Log(Price)', fontsize=15)

# Highlight halving events
for halving_date in halving_dates:
    plt.axvline(pd.to_datetime(halving_date), color='red', linestyle='--', label=f'Halving {halving_date}')
plt.legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'Year' and count the number of elements (rows) in each year
count_by_year_df1 = df1.groupby('Year').size()
count_by_year_df2 = df2.groupby('Year').size()
count_by_year_df3 = df3.groupby('Year').size()

# Combine the data into a single dataframe for easier plotting
combined_df = pd.DataFrame({
    'Table 1': count_by_year_df1,
    'Table 2': count_by_year_df2,
    'Table 3': count_by_year_df3
}).fillna(0)  # Fill NaN with 0 for years missing in any table

# Calculate the differences between the tables
combined_df['Discarded by C2'] = combined_df['Table 2'] - combined_df['Table 3']
combined_df['Discarded by C1'] = combined_df['Table 1'] - combined_df['Table 2']

# Plot the stacked histogram
combined_df[['Table 3', 'Discarded by C2', 'Discarded by C1']].plot(
    kind='bar', stacked=True, color=['green', 'orange', 'skyblue'], figsize=(10, 6))

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Number of Records')
plt.title('Number of Included and Discarded Records per Year')
plt.legend(['Included', 'Discarded by C2', 'Discarded by C1'], title='Records')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a new DataFrame to hold the occurrences
occurrences = pd.DataFrame(index=range(df['Start_Year'].min(), df['End_Year'].max() + 1),
                           columns=range(1, 13)).fillna(0)

# Iterate through the rows of the original DataFrame and update occurrences
for index, row in df.iterrows():
    start_year = row['Start_Year']
    end_year = row['End_Year']
    start_month = row['Start_Month']
    end_month = row['End_Month']

    # If the start and end years are the same, update only for that year
    if start_year == end_year:
        occurrences.loc[start_year, start_month:end_month] += 1
    else:
        # Update for start year
        occurrences.loc[start_year, start_month:] += 1
        # Update for end year
        occurrences.loc[end_year, :end_month] += 1
        # Update for years in between
        for year in range(start_year + 1, end_year):
            occurrences.loc[year, :] += 1

# Plotting
plt.figure(figsize=(10, 6))
plt.imshow(occurrences, cmap='viridis', aspect='auto')
plt.colorbar(label='Occurrences')
plt.title('Most Studied Interval of Times')
plt.xlabel('Month')
plt.ylabel('Year')
plt.xticks(range(12), range(1, 13))
plt.yticks(range(len(occurrences.index)), occurrences.index)
plt.show()
```
